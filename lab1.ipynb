```python
import pandas as pd

dataset = pd.read_csv('resources/mobile.csv', encoding='windows-1251', sep='\t')
```

```python
# посмотрим на первые 10 строк
dataset.head(10)
```

```python
# столбец "код" нам не нужен
dataset = dataset.drop(columns=['Код'])
dataset.describe()
```

```python
# построим матрицу корреляций
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(dataset.corr(),annot=True,cmap='RdYlGn',linewidths=0.3)
fig=plt.gcf()
fig.set_size_inches(12,12)
```

```python
# смотрим зависимость количества смс от возраста
x = dataset['Возраст'].values
y = dataset['Количество SMS за месяц'].values

plt.xlabel('Возраст')
plt.ylabel('Количество SMS за месяц')
plt.title('Точечный график зависимости возраста от количества смс')
plt.grid(True)
plt.scatter(x, y, marker='.')
```

```python
# смотрим зависимость расхода от продолжительнорсти разговора
x = dataset['Средняя продолжительность разговоров'].values
y = dataset['Среднемесячный расход'].values

plt.xlabel('Средняя продолжительность разговоров')
plt.ylabel('Среднемесячный расход')
plt.title('Точечный график зависимости расхода от продолжительности')
plt.grid(True)
plt.scatter(x, y, marker='.')
```
